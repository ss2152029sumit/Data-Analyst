```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example DataFrames (Replace these with SQL query results)
data_revenue = {'Customer': ['John Doe', 'Jane Smith', 'Alice Johnson'], 'Total Spent': [1350.50, 800.00, 410.00]}
data_sales = {'Category': ['Electronics', 'Furniture', 'Accessories'], 'Total Sold': [75, 10, 100]}
data_status = {'Status': ['Completed', 'Pending', 'Cancelled'], 'Total Orders': [2, 1, 1]}
data_ratings = {'Category': ['Electronics', 'Furniture', 'Accessories'], 'Average Rating': [4.5, 4.0, 4.2]}
data_age_groups = {'Age Group': ['Under 18', '18-30', '31-50', '51+'], 'Total Customers': [0, 3, 1, 1]}

# Convert to DataFrames
df_revenue = pd.DataFrame(data_revenue)
df_sales = pd.DataFrame(data_sales)
df_status = pd.DataFrame(data_status)
df_ratings = pd.DataFrame(data_ratings)
df_age_groups = pd.DataFrame(data_age_groups)

# Visualization Examples
# Revenue by Customer
plt.figure(figsize=(8, 6))
sns.barplot(x='Customer', y='Total Spent', data=df_revenue, palette='viridis')
plt.title('Revenue by Customer')
plt.show()

# Sales by Product Category
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Total Sold', data=df_sales, palette='coolwarm')
plt.title('Sales by Product Category')
plt.show()

# Order Status Distribution
plt.figure(figsize=(8, 6))
sns.barplot(x='Status', y='Total Orders', data=df_status, palette='Set2')
plt.title('Order Status Distribution')
plt.show()

# Average Ratings by Category
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Average Rating', data=df_ratings, palette='Blues')
plt.title('Average Ratings by Category')
plt.show()

# Customer Age Groups
plt.figure(figsize=(8, 6))
sns.barplot(x='Age Group', y='Total Customers', data=df_age_groups, palette='Greens')
plt.title('Customer Age Groups')
plt.show()

# Box Plot for Revenue Distribution
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_revenue['Total Spent'], color='purple')
plt.title('Revenue Distribution')
plt.xlabel('Total Spent')
plt.show()

# Correlation Heatmap
numeric_data = pd.DataFrame({
    'Total Spent': df_revenue['Total Spent'],
    'Total Sold': df_sales['Total Sold'],
    'Average Rating': df_ratings['Average Rating']
})
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Order Status Distribution (Donut Chart)
plt.figure(figsize=(8, 6))
explode = (0.1, 0, 0)  # Explode the first slice
plt.pie(df_status['Total Orders'], labels=df_status['Status'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'), explode=explode)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Order Status Distribution')
plt.show()

# Customer Age Distribution
plt.figure(figsize=(8, 6))
sns.barplot(x='Age Group', y='Total Customers', data=df_age_groups, palette='mako')
plt.title('Customer Age Distribution')
plt.xlabel('Age Group')
plt.ylabel('Total Customers')
plt.show()

# Average Ratings by Category
plt.figure(figsize=(8, 6))
sns.barplot(y='Category', x='Average Rating', data=df_ratings, palette='Blues')
plt.title('Average Ratings by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()

# Revenue vs. Sales
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=df_revenue['Customer'], y=df_revenue['Total Spent'], color='blue', label='Revenue', ax=ax)
sns.barplot(x=df_sales['Category'], y=df_sales['Total Sold'], color='orange', label='Sales', ax=ax)
ax.set_title('Revenue vs. Sales')
ax.legend()
plt.show()



```
